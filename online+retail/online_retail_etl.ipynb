```python
def get_transaction_type(row):
    if row['Quantity'] < 0:
        return 'Retour'
    elif row['UnitPrice'] == 0.0:
        return 'Gratuit'
    else:
        return 'Achat'

df['type_transaction'] = df.apply(get_transaction_type, axis=1)
df['type_transaction'].value_counts()
df.head()
```

```python
# Détection des valeurs aberrantes pour les colonnes quantitatives principales
# Ici, nous utilisons lIQR pour détecter les outliers sur 'Quantity' et 'UnitPrice'

def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

outliers_quantity = detect_outliers_iqr(df, 'Quantity')
outliers_unitprice = detect_outliers_iqr(df, 'UnitPrice')

print(f"Nombre de valeurs aberrantes pour Quantity : {outliers_quantity.shape[0]}")
print(f"Nombre de valeurs aberrantes pour UnitPrice : {outliers_unitprice.shape[0]}")
```
